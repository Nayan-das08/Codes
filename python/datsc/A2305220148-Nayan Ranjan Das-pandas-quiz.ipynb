## 3. difference between `iloc` and `loc` with examples

when we use `iloc` to display/select rows and columns, the output doesn't include the last index

also, we can use numeric indices for columns

```python
market_df.iloc[1:3, :2]
```

when using `loc`, the output includes the last index mentioned for rows.

also, we need to specify the columns by the specified name to access them

```python
market_df.loc[1:3, ['Ord_id', 'Prod_id']]
```
