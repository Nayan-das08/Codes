```python
new = pd.merge(market, cust, on="Cust_id")
new.shape
```

```python
new.loc[new.Customer_Segment == "CORPORATE", ['Ord_id', 'Sales']]
```

## select from above dataframe where customer segment is home office and profit is greater than 500

```python
new.loc[(new.Customer_Segment == "HOME OFFICE") & (new.Profit > 500)]
```

## Show rows where customer name is Alex

```python
new.loc[(new.Customer_Name == "ALEX AVILA") & (new.Profit > 500) & (new.Customer_Segment == "HOME OFFICE")]
```

## Sales greater than or equal to 300 and profit greater than 1000

```python
new.loc[(new.Sales >= 300) & (new.Profit > 1000)]
```
