```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

# Pre-process Data

```python
df = pd.read_csv('Housing.csv')
df.head()
```

```python
df.shape
```

```python
True in df.duplicated()
```

```python
categorical = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']
numeric = ['price', 'area', 'bedrooms', 'bathrooms', 'stories']
```

```python
sns.pairplot(data=df)
```

```python
sns.heatmap(df.corr(), annot=True)
```

```python
# plt.figure(figsize=(20,12))
# plt.subplot(2,3,1)
for i in categorical:
    plt.figure()
    sns.boxplot(x=df[i], y=df.price, data=df)
```

```python
df[categorical] = df[categorical].apply(LabelEncoder().fit_transform)
df.head()
```

```python
status = pd.get_dummies(df.furnishingstatus, drop_first=True)
status.head()
```

```python
df2 = pd.concat([df, status], axis=1)
df2.drop(columns=['furnishingstatus'], inplace=True)
df2.head()
```

```python
scaler = MinMaxScaler()
df_minmax = df2.copy()
df_minmax[numeric] = scaler.fit_transform(df2[numeric])
df_minmax.head()
```

# split data

```python
features = list(df_minmax.columns)
features.pop(0)
features
```

```python
x_train, x_test, y_train, y_test = train_test_split(df_minmax[features], df_minmax.price, train_size=0.7, test_size=0.3, random_state=100)
```

# set model

```python
import statsmodels.api as sm
```

```python
x_train_lm = sm.add_constant(x_train[['area', 'bathrooms', 'airconditioning', 'stories', 'parking', 'prefarea', 'mainroad', 'guestroom', 'basement', 'hotwaterheating', 'semi-furnished']])
lr = sm.OLS(y_train, x_train_lm).fit()
lr.rsquared
```

```python
lr.summary()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif = pd.DataFrame()
vif['Features'] = x_train.columns
vif['VIF'] = [variance_inflation_factor(x_train.values, i) for i in range(x_train.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```
