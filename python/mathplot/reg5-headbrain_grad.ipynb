```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import regression as reg
```

```python
data = pd.read_csv('headbrain.csv')
data = data.dropna()
data.head(n=2)
```

```python
#x = data['Avg. Area Income']
#y = data['Price']
x = data['Brain Weight(grams)']
y = data['Head Size(cm^3)']
```

```python
A = B = 0       # initial assumption
L = 0.00000001     # learning rate (needs to be really small)
A_list = []
B_list = []
cost_list = []

for i in range(500):
    y_pred = A + B*x
    cost = np.sum((y-y_pred)**2) / len(x)
    A_change = np.sum(y-y_pred) * (-2/len(x))
    B_change = np.sum((y-y_pred)*x) * (-2/len(x))
    A = A - L*A_change
    B = B - L*B_change
    cost_list.append(cost)
    A_list.append(A)
    B_list.append(B)
print(f"{i} \t A = {'%.7f'%A} \t\t B = {'%.7f'%B}   \t cost = {cost}")

plt.scatter(x,y)
plt.plot(x,y_pred,color='r')
plt.xlabel('Brain Weight (grams)')
plt.ylabel('Head Size (cm^3)')
plt.title('Regression Line')
```

```python
plt.scatter(A_list, cost_list)
plt.title('Plot of Gradient Descent for Co-efficient A')
plt.xlabel('Co-efficient A')
plt.ylabel('Cost Function')
```

```python
plt.scatter(B_list, cost_list)
plt.title('Plot of Gradient Descent for Co-efficient B')
plt.xlabel('Co-efficient B')
plt.ylabel('Cost Function')
```
