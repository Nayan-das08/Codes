```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import regression as reg
```

```python
df = pd.read_csv('headbrain.csv')
df.head()
```

```python
df.plot(kind='scatter', x='Brain Weight(grams)', y='Head Size(cm^3)')
```

```python
x_train, x_test, y_train, y_test = train_test_split(df['Brain Weight(grams)'], df['Head Size(cm^3)'])
```

```python
plt.scatter(x_train,y_train, label='train', color='r')
plt.scatter(x_test,y_test, label='test', color='b')
plt.legend()
```

```python
x_train.values.reshape(-1,1)
```

```python
lr = LinearRegression()
lr.fit(x_train.values.reshape(-1,1), y_train)
```

```python
pred = lr.predict(x_test.values.reshape(-1,1))
```

```python
plt.plot(x_test,pred, label='regression', color='r')
plt.scatter(x_test,y_test, label='test data')
plt.legend()
```
