```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('datasets/CarPrice_Assignment.csv')
df.head()
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(df.corr(), annot=True)
```

```python
df.columns
```

```python
col = ['enginesize', 'curbweight', 'horsepower', 'carwidth']
```

```python
X = df.loc[:, col]
X.head()
```

```python
y = df.price
m = X.shape[0]
n = X.shape[1]
```

```python
w = np.random.rand(n)
b = np.random.rand()

cost_list = []
w_list = []
b_list = []
L = 0.0000001
epochs = 500
dw = np.zeros(n)

for i in range(epochs):
    y_pred = np.dot(X,w)+b
    cost = np.sum((y_pred-y)**2)/(2*m)
    for j in range(n):
        dw[j] = np.dot((y_pred-y),X.values[:,j])/m
    db = np.sum((y_pred-y))/m
    w = w - L*dw
    b = b - L*db
    cost_list.append(cost)
    w_list.append(w)
    b_list.append(b)
#     if i%(epochs//10) == 0:
print(f'i={i} \tcost={cost:.4f}')

# plt.subplot(2,1,1)
plt.figure()
plt.plot(cost_list);

# plt.subplot(2,1,2)
# plt.figure()
# plt.scatter(X.values[:,0],y)
# plt.plot(X.values[:,0],y_pred,color='r');

RSS = np.sum((y_pred-y)**2)
TSS = np.sum((y_pred-np.mean(y))**2)
1-(RSS/TSS)
# R2
```
