```python
col = ['enginesize', 'curbweight', 'price']
```

```python
data = df.loc[:,col]
data.head()
```

```python
X = data.drop(columns='price')
y = data.price
n = X.shape[1]
m = X.shape[0]
```

```python
# w_init = np.random.rand(n)
# b_init = np.random.rand()
# w = w_init
# b = b_init
w = np.array([0,0])
b = 0
```

```python
def f(w,b,x):
    return np.dot(x,w) + b
```

```python
def compute_cost(w, b, X, y):
    y_pred = f(w,b,X.values)
    y_actual = y
    sum = np.sum(y_pred-y_actual)**2
    return sum/(2*m)
```

```python
def compute_gradient(X,y,w,b):
# this function will loop m times (for each sample)
# calculate the error
# then loop n times (for each feature) to get gradient for each feature
# then get gradient for bias
# return the gradients
    dw = np.zeros(w.shape)
    db = 0
    for i in range(m):
        error = np.dot(X.values[i],w)+b - y[i]
        for j in range(n):
            dw[j] += error*X.values[i,j]
        db += error
    return dw/m, db/m

# compute_gradient(X,y)
```

```python
def gradient_descent(X, y, w, b, alpha, epochs):
    cost_list = []
    
    for i in range(epochs):
        dw, db = compute_gradient(X,y,w,b)
        w = w - alpha*dw
        b = b - alpha*db
        cost_list.append(compute_cost(w,b,X,y))
        if i%10 == 0:
            print(f'epoch {i:3d}:, \tcost= {cost_list[-1]:8.2f}')
    
    return cost_list, w, b
```

```python
J, new_w, new_b = gradient_descent(X, y, w, b, alpha=1, epochs=100)
```

```python
J, new_w, new_b = gradient_descent(X, y, w, b, alpha=100, epochs=100)
```
